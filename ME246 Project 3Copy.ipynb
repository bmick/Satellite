```python
def VocCurveFit():
    # This code solves Task 1, it finds a linear curve fit for 
    # The data given on the handout. 
    
    T_vals = [30.,42.,50.,60.,82.]
    T_vals = [T+273 for T in T_vals] # Had to make it Kelvin
    V_vals = [.37,.35,.33,.31,.27]
    A,B = np.polyfit(T_vals, V_vals,1)
    if plots:
        plt.plot(T_vals,V_vals,'g*')
        plt.plot([293,373],[A*T+B for T in [293,373]],'b:')
        plt.xlabel("Temperature, K")
        plt.ylabel("Voltage, V")
        plt.title("$V_{oc}$ vs Temp")
        plt.show()
    
    return A,B
```

```python
def IoIvfromT(T):
    if debug:
        print "Entering IoIv from T function"
        print "T = ", T
    #Task 1
    #This code returns Io/Iv as a function of T
    #Based on the linear temperature fit for Voc
    #As given in VocCurveFit
    A,B = VocCurveFit()
    Iv_Io = np.exp((A*T + B)*q_e/(K_b*T))-1
    if debug:
        print "Iv/Io = {:.3e}".format(Iv_Io)
    Io_Iv = 1/Iv_Io
    if debug:
        print "In function, Io/Iv = ", Io_Iv
    return Io_Iv
```

```python
def integrand(x):
    #Helper function for task 3
    return x**2/(np.exp(x)-1)
```

```python
def PowerfromRLpv(R_Lpv):
    # Task 3
    if debug:
        print "\n Entering PowerfromRLpv function"
    phi = P/(math.pi**4*K_b*T_source/(2.404*15.)) # from 14.16
    # print "PowerfromRLPV code Power = ",P
    X = q_e*V_g/(K_b*T_source)
    Val, err = quad(integrand, X, 40.)
    phi_g = phi*.416*Val
    if debug:
        print "PowerfromRLpv Io_Iv value = ",Io_Iv
    global Iv
    Iv = phi_g*q_e*A_pv
    global Io
    Io = Io_Iv*Iv
    
    if debug:
        print "Io/Iv = ", Io_Iv
        print "Iv = ", Iv
    if debug:
        print "Io = ", Io
    
    if plots:
        IL = np.arange(64.)
        V_L = [K_b*T_pv/q_e*math.log((Io_Iv)**-1-I/Io+1)-I*Rs for I in IL]
        V_L2 = [R_Lpv*I for I in IL]
        IL = np.arange(len(V_L))
        plt.plot(IL,V_L,IL,V_L2)
        plt.show()

    #if debug:
    i = 10.
    b = Iv+Io-i
    #print "B = ", b, "Error (B) = ",f_VL(b)

    while f_VL(b)>0:
        i = i/10
        b = Iv+Io-i
        #print "i = ", i, "Error (B) = ",f_VL(b)
        if i < 1e-100:
            break
    
    #print "Final B = ", b, "Final Error (B) = ",f_VL(b)
    
    I_1 = brentq(f_VL,1,b)
    
    if debug:
        print "Current = ", I_1, "Voltage = ",I_1*R_Lpv
    P_Lpv = I_1**2*R_Lpv
    if debug:
        print "PV Power = ",P_Lpv
    eff_cell = P_Lpv/(P*A_pv)
    
    return I_1,I_1*R_Lpv, P_Lpv, eff_cell
```

```python
def f_VL(IL):
    try:
        E = K_b*T_pv/q_e*math.log((Io_Iv)**-1-IL/Io+1)-IL*Rs-IL*R_Lpv
    except ValueError:
        print "Value Error raised"
        print "Io/Iv = ", Io_Iv
        print "IL = ", IL
        print "Io = ", Io
        print "Log Insides = ",(Io_Iv)**-1-IL/Io+1
        # print "Photovoltaic Current Solver"
    return E
```

```python
def f_Ite(IL):
    debug = False
    if debug:
        print "Thermoelectric Current Solver"

    if debug:
        print "n = ", n
        print "Seebeck = ", seebeck
        print "IL = ", IL
        print "T_H = ", T_H
        print "cond_batt = ", cond_batt
        print "T_C = ", T_C
        print "R_batt = ", R_batt
        print "Qh = ", Qh
    E = n*seebeck*IL*T_H + cond_batt*(T_H-T_C)-.5*IL**2*R_batt - Qh
    return E
```

```python
def TEModule(T_H):
    #From Task 4 Description. I'll be honest, I don't really understand this.
    #First, Sequentially compute the parameters listed below:
    debug = False
    
    #a The Value of (Thermal Conductance * Electrical Resistance)min
    RestCondMin = ((rest_a*cond_a)**.5 + (rest_b*cond_b)**.5)**2
    if debug:
        print "(Restistance * Conductance)min = ",RestCondMin
    
    #b The value of Z based on R*C min
    Z = seebeck**2/RestCondMin
    if debug:
        print "The Figure of Merit = {:.4f}".format(Z)
    
    # We need to guess T_C here: 
    #i Guess TC:
    global T_C
    T_C = T_sat + 1 # Pretty much just picking a trivial, reasonable guess 
    
    iterations = 0
    Err_q = 10

    while abs(Err_q) > .1:
        if debug:
            print "\nIteration {:.0f}:".format(iterations)
            print "Current T_C guess = {:.2f} K".format(T_C)
        
        #c The value of m_max eff (This is where it gets sticky, m requires T_H and T_C, we don't know T_C)
        m_maxeff = (1+.5*(T_H+T_C)*Z)**.5
        if debug:
            print "m for max efficiency = {:.2f}".format(m_maxeff)
        
        #d the value of Thermoelectric Load Resistance / number of pairs * m(max eff)
        global R
        R = RL_te/(n*m_maxeff)
        global R_batt
        R_batt = R*n #total resistance of the battery
        if debug:
            print "Battery Resistance = {:.4f} Ohms".format(R_batt)
        
        #e the value of conductance = (thermal conductance * electrical resistance)min / electrical resistance
        global cond
        cond = RestCondMin/R
        global cond_batt
        cond_batt = cond*n #total thermal conductance of the battery
        if debug:
            print "Battery Thermal Conductance = {:.4f} W/K".format(cond_batt)
        
        #Then iteratively compute TC and QH required to satisfy eqn (4), (10), and (11)

        #ii use eqn (4) to calculate (Qc):
        Qc = kw_eff*(T_C-T_sat)*A_pv/tw
        if debug:
            print "Waste Heat Rejected, Qc = {:.2f} W".format(Qc)
        
        #iii compute efficiency using guessed TC value:
        eta = ((T_H-T_C)/(T_H))*((1+m_maxeff)**2/m_maxeff*(Z*T_H)**-1 + 1 + 1/(2*m_maxeff)*(1+T_C/T_H))**-1
        if debug:
            print "Thermoelectric Efficiency = {:.4%}".format(eta)

        #iv compute the power from eqn (10):
        
        Power_out = (n**2*seebeck**2*(T_H-T_C)**2*RL_te)/(RL_te+R_batt)**2 
        #Here we are using the R calculated from the 
        # max efficiency up in step 4, it's what Carey said to do when I asked him. Something about designing the TE
        # device to match the load (clearly we want to be at max eff with our design)
        if debug:
            print "n = ", n
            print "seebeck = ", seebeck
            print "T_H = ", T_H
            print "T_C = ", T_C
            print "RL_te = ",RL_te
            print "R_batt = ", R_batt
            print "Power out of the TE cell = {:.2f} W".format(Power_out)

        #v Compute the heat input and rejection for the thermoelectric device
        global Qh
        Qh = Power_out/eta
        Qc_te = Qh - Power_out

        if debug:
            print "Q input, needed for TE Power = {:.2f} W".format(Qh)
            print "Q rejected, uneeded for TE Power = {:.2f} W".format(Qc_te)
            print "Qc from TC & T_Sat = {:.2f} W".format(Qc)

        #vi Compute the error Err_q as Qc_te - Qboil (where Qboil = Qc)
        Err_q = Qc_te - Qc
        if debug:
            print "Error = {:.2f}".format(Err_q)
                
        T_C = (Qc_te)*tw/(kw_eff*A_pv) + T_sat
        Qc = Qc_te
        
        iterations += 1
        if iterations > 10:
            print "Didn't converge"
            break
    
    if debug:
        print "\nCode Converged. Entering Newton-Raphson Solver for IL_te:"
        
    #Newton Raphson solver from above:
    I_0 = 0
    I_1 = 1
    iterations = 0
    while abs(I_0-I_1)>.01:
        iterations += 1
        if debug:
            print "\nIteration {:.0f}:".format(iterations)
        I_0 = I_1
        I_1 = I_0 - f_Ite(I_0+1)/(f_Ite(I_0+1)-f_Ite(I_0))
        if debug:
            print "Old Guess for IL_te = {:.2f} A".format(I_0)
            print "New guess for IL_te = {:.2f} A".format(I_1)
        if iterations > 10:
            break
    I_te = I_1
    V_te = Power_out/I_te

    return T_C, Qh, Qc, I_te, V_te
```
